```python
def car(input_x):
    if re.match(r'土地\d+建物\d+車位1$', input_x):
        return 1
    elif re.match(r'土地\d+建物\d+車位2$', input_x):
        return 2
    elif re.match(r'土地\d建物\d車位3$', input_x):
        return 3
    elif re.match(r'土地\d建物\d車位4$', input_x):
        return 4
    elif re.match(r'土地\d建物\d車位5$', input_x):
        return 5
    elif re.match(r'土地\d建物\d車位6$', input_x):
        return 6
    elif re.match(r'土地\d建物\d車位7$', input_x):
        return 7
    elif re.match(r'土地\d建物\d車位9$', input_x):
        return 9
    elif re.match(r'土地\d建物\d車位10$', input_x):
        return 10
    elif re.match(r'土地\d建物\d車位12$', input_x):
        return 12
    elif re.match(r'土地\d建物\d\d車位13$', input_x):
        return 13
    elif re.match(r'土地\d建物\d\d車位38$', input_x):
        return 38
    elif re.match(r'土地\d建物\d車位44$', input_x):
        return 44
    else:
        return 0
df_all['交易筆棟數'] = df_all['交易筆棟數'][1:8865].apply(car)
df_all
```
